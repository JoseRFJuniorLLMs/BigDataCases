```python
#Arquivo = GOTS01.json
#Temp - S01, S02, S03

def tratamento(p1,p2):
    arquivo = p1
    temporada = p2
    
    with open(arquivo) as arq:    
        data = json.load(arq)
    
        GOTA = pd.DataFrame(data['Episodes'])

        GOTA[ 'imdbRating' ] = GOTA['imdbRating'].astype(float)

        GOTB = GOTA.assign(season = temporada).copy(deep=True)

        return GOTB
```

```python
GOT = GOT.append(tratamento('GOTS03.json','S03'), ignore_index=True)
```

```python
GOT = GOT.append(tratamento('GOTS04.json','S04'), ignore_index=True)
```
