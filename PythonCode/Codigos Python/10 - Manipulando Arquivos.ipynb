```python
#Strip - Funao que retira espaos, equivalente ao TRIM
frase = '   Frase com espaos   '
print(frase)
print(frase.strip())
```

```python
# importante saber que o PYTHON l um arquivo como uma sequencia de caracteres e identifica 
#automaticamente o /n que  o comando new line, ou seja, o ENTER em ASCII

lista = ['palavra1','palavra2']  #Criando uma variavel lista

with open("linhas.csv") as f: #abrindo o arquivo
    for linha in f: #realizando um for no arquivo
        lista.append(linha.strip()) #Adicionando uma nova linha ao fim da lista, removendo espaos
        print(lista)
```
